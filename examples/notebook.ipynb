# 🧠 Mixture of Experts - Interactive Tutorial

This notebook guides you through the MoE system step by step.

## 📦 Setup

```python
import sys
sys.path.insert(0, '..')

from src.core.config import MoEConfig
from src.core.state import create_initial_state
from src.graph.builder import MoEGraphBuilder
import os

# Set your API key
os.environ['GROQ_API_KEY'] = 'your_key_here'
```

## 🔧 Initialize the System

```python
# Create configuration
config = MoEConfig(groq_api_key=os.getenv('GROQ_API_KEY'))

# Build the graph
builder = MoEGraphBuilder(config)
graph = builder.build()

print("✅ System initialized!")
print(f"Available experts: {list(config.expert_configs.keys())}")
```

## 🎯 Test 1: Single Expert

```python
query = "What is Python?"

state = create_initial_state(query)
result = graph.invoke(state)

print(f"Query: {query}")
print(f"Selected Experts: {result['selected_experts']}")
print(f"\nFinal Answer:\n{result['final_answer']}")
```

## 🎯 Test 2: Multiple Experts

```python
query = "Explain recursion with a creative analogy"

state = create_initial_state(query)
result = graph.invoke(state)

print(f"Query: {query}")
print(f"Selected Experts: {result['selected_experts']}")

print("\nExpert Responses:")
for expert, response in result['expert_responses'].items():
    print(f"\n{expert.upper()}:")
    print(response[:200] + "...")

print(f"\n\nSynthesized Answer:\n{result['final_answer']}")
```

## 📊 Visualize the Network

```python
from ui.components.visualization import MoEVisualizer

visualizer = MoEVisualizer()

# Create network graph
fig = visualizer.create_network_graph(
    result['selected_experts'],
    result['expert_responses']
)

fig.show()
```

## 📊 Confidence Scores

```python
fig = visualizer.create_confidence_chart(result['confidence_scores'])
fig.show()
```
